## Descripción del Dataset

Las instituciones de educación superior registran una cantidad importante de datos sobre sus estudiantes, lo que representa un potencial considerable para generar información, conocimiento y seguimiento. Tanto la deserción escolar como el fracaso educativo en la educación superior son un obstáculo para el crecimiento económico, el empleo, la competitividad y la productividad, impactando directamente en la vida de los estudiantes y sus familias, las instituciones de educación superior y la sociedad en su conjunto.

# Historia

El éxito académico en la educación superior es vital para el empleo, la justicia social y el crecimiento económico. La deserción es en mayor inconveniente que las instituciones de educación superior deben abordar para mejorar su éxito. Según el informe de la Comisión Europea, son muchos los estudiantes que abandonan antes de finalizar sus estudios de educación superior. Incluso en el país más exitoso (Dinamarca), solo alrededor del 80% de los estudiantes completan sus estudios, mientras que en Italia, esta tasa es solo del 46%. Este informe destaca los factores clave que llevan a los estudiantes a abandonar los estudios, siendo la principal razon las condiciones socioeconómicas de cada país e institucion en particular.
El conjunto de datos incluye datos demográficos, datos socioeconómicos y macroeconómicos, datos al momento de la inscripción de los estudiantes y datos al final del primer y segundo semestre. Los datos se refieren a expedientes de alumnos matriculados entre los cursos académicos 2008-2009 al 2018-2019. Estos incluyen datos de carreras de grado de diferentes campos del conocimiento, como agronomía, diseño, educación, enfermería, periodismo, administración, servicio social y tecnologías.
